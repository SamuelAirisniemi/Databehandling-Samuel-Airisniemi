```python
fig, axes = plt.subplots(2, 2)

college["Apps"].plot.hist(bins=10, ax=axes[0,0])
axes[0,0].set_title("Applications (10 bins)")

college["Accept"].plot.hist(bins=20, ax=axes[0,1])
axes[0,1].set_title("Acceptances (20 bins)")

college["Enroll"].plot.hist(bins=30, ax=axes[1,0])
axes[1,0].set_title("Enrollments (30 bins)")

college["Top10perc"].plot.hist(bins=40, ax=axes[1,1])
axes[1,1].set_title("Top 10% (40 bins)")

plt.tight_layout()
plt.show()
```

```python
college["Elite"] = pd.cut(college["Top10perc"], bins = [0, 50, 100], labels = ["No", "Yes"])

fig, axes = plt.subplots(1, 2, figsize = (16, 6))

sns.scatterplot(x = "Personal", y = "Expend", hue = "Elite", ax = axes[0], data = college)

axes[0].set_title("Instructional vs Personal Spending by Elite Status")
axes[0].set_xlabel("Personal Spending")
axes[0].set_ylabel("Instructional Spending")

melted = college.melt(id_vars = "Elite", value_vars = ["Expend", "Personal"])
sns.boxplot(data = melted, x = "variable", y = "value", hue = "Elite", ax = axes[1])
axes[1].set_title("Spending Distributions by Elite Status")
axes[1].set_xlabel("Spending Category")
axes[1].set_ylabel("Amount")

plt.tight_layout()
plt.show()

#plt.figure(figsize = (10, 6))
#sns.scatterplot(data=college, x = "Expend", y = "Personal", hue = "Elite")
#plt.title("Instructional Spending vs Personal Spending by Elite Status")
#plt.xlabel("Instructional Spending")
#plt.ylabel("Personal Spending")
#plt.legend(title="Elite Status")
#plt.show()
```

Fråga: Finns det ett samband mellan personliga utgifter och elitstatus?

Här undersöker vi en jämförelse av utbildningsutgifter och personliga utgifter beroende om man läser elituniversitet eller inte. Vi kan se ett tydligt mönster där elituniversitet spenderar betydligt mycket mer på undervisning. Det finns också en korrelation mellan "Expend" och "Personal", vilket betyder att skolor som investerar mer i undervisning ofta har högre personliga utgifter. I boxplot grafen ser vi även att elituniversitet spenderar mycket mer än icke-elituniversitet. Nästan 2-3x så mycket.

Vi ser även en överlappning mellan elit och icke-elit. Vilket tyder på att även om undervisningsutgifter är en stark differentieringsfaktor så påverkas personliga utgifter också av andra faktorer (plats, elevdemografi, leveranskostnad). Elituniversitet är starkt kopplad till undervisningsutgifter. Vilket vi kan anta eftersom elituniversitet ofta har större anslag och har råd med högre lärarlöner, mindre klasser och mer resurser. Personliga utgifter är mindre starkt kopplad till elituniversitet men trenden är fortfarande uppåtgående. Studenter på elituniversitet kan ha högre levnadskostnader och mer valfria utgifter.
