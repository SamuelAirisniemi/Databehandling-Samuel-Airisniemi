```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

olympics = pd.read_csv("C:/Users/samue/Documents/code/Databehandling-Samuel-Airisniemi/athlete_events.csv")
germany_medals = olympics[(olympics["NOC"] == "GER")] & olympics["Medal"].notna()

medals_per_year = germany_medals.groupby("Year")["Medal"].count().reset_index()
medals_per_year = medals_per_year.sort_values("Year")
plt.figure(figsize=(12,6))
sns.barplot(data=medals_per_year, x="Year", y="Medal", color="steelblue")
plt.title("Antal medaljer per OS för Tyskland")
plt.ylabel("Antal medaljer")
plt.xlabel("År")
plt.xticks(rotation=45)
plt.show()
```
